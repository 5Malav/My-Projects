```python
X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]\

y=df['total_bill']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,\
                        random_state=10)
```

```python
X_train.head()
```

```python
#feature encoding(label encoding and onhot encoding)

#there will always be problem of data leakage. so first divide
#data and then covnvert
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python
X_train.head(10)
```

```python
X_test['sex']=le1.transform(X_test['sex'])
X_test['smoker']=le2.transform(X_test['smoker'])
X_test['time']=le3.transform(X_test['time'])
```

```python
X_test.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#convert day record in numbers

ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],\
                remainder='passthrough')

#drop=first means 1st column will be dropped and remaining
#columns will be converted

#3rd columne day will be converted and [] means reshaping data
#ow it will show error to reshape data
```

```python
#

ct.fit_transform(X_train)
```

```python
import sys
import numpy as np

np.set_printoptions(threshold=sys.maxsize)

X_train=ct.fit_transform(X_train)
```

```python

X_test=ct.fit_transform(X_test)
```

```python
X_test
```

```python
#apply support vector regression

from sklearn.svm import SVR

svr=SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error

print(r2_score(y_test,y_pred))

print(mean_absolute_error(y_test,y_pred))
```

```python
#Hyper Parameter turning using GridSearch CV

from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3)

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error

print(r2_score(y_test,grid_pred))

print(mean_absolute_error(y_test,grid_pred))
```
